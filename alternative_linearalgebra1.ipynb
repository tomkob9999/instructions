# Box Reshape

To start, there is a plain square box, sitting flat.





Let's reshape the box.  Only one rule is that the bottom left corner is fixed at a point.  Otherwise as long as it is a box,
* It can rotate
* It can stretch vertically and horizontally







Stricly speaking during the reshaping, it needs to remain a parallelgram.
* Lines need remain straight
* Parallel lines need to remain parallel

For this reshaping, you need only 4 parameters to specify the exact reshaping.  By giving Ax, Ay, Bx, By, you can reconstruct the reshaped box.



Then, you can create a net by laying multiple boxes aligned side by side.  It is quite obvious so I dont think it needs any explanation.



Back to the single box...

The original box with the fixed botton left corner can be reshaped by specifying x and y of top left corner and bottom right corner.

It can also reshaped by giving below 4 parameters.

* Stretching rate of vertical line
* Stretching percentage of horizontal line
* New angle of vertical line
* New angle of horizontal line




Together with the reshaping of the box and the net creation out of a single box, you can reshape from a net to a new net.

I will demonstrate a progam that converts into a net to another net.


```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

def normalize_vector(vector):
    magnitude = norm_of(vector)
    if magnitude == 0:
        return [0, 0]
    return vector / magnitude

def norm_of(vec):
    return np.sqrt(vec[0]**2+vec[1]**2)

def equals_more_or_less(a, b):
    return a > b - 0.00001 and a < b + 0.00001

def convToLists(npm):
    return [list(n) for n in npm]

def convToListsWithRound(npm, r):
    return [[np.round(nn, r) for nn in n] for n in npm]

print("")
print("*********************************************************")
print("* CELLS ARE INDICATED AS BELOW IN EQUATIONS:")
print("* [tl (Top Left)     tr (Top Right)]")
print("* [bl (Bottom Left)  br (Bottom Right)]")
print("*********************************************************")
print("")
print("")
print("")


lislis = []
# Define the linear transformation matrix
# Squeezing matrix, no rotation
# [[x, y], [y, x]] types only squeezes without rotation
lis = [[[1-a, a], [a, 1-a]]for a in list(np.linspace(0, 1, 11))]
lislis.append(lis)
# Square matrix Rotation
# [[x, -y], [y, x]] (negative can be any one of them) types only squeezes without rotation
lis = [[[1-a, a], [-a, 1-a]]for a in list(np.linspace(0, 1, 11))]
# lis = [[[(1-a)*2, a], [a, 1-a]]for a in list(np.linspace(0, 1, 11))]
# lis = [[[(1-a), a], [a, 1-a]]for a in list(np.linspace(0, 1, 11))]
lislis.append(lis)

# Variations of the same eigen space
lis = [[[8+i, 2], [4, 1+i]] for i in range(-10, 11, 1)]
# lis = [[[8, 2+i], [4+i, 1]] for i in range(-10, 11, 1)]
# lis = [[[4+i, 1], [8, 2+i]] for i in range(-10, 11, 1)]
# lis = [[[4+i, 1], [8, 2+i]] for i in range(-10000, 10001, 1000)]
lislis.append(lis)

# Full patterns if 0, -1, 1
# values = [0, -1, 1]
# combinations = list(product(values, repeat=4))
# lis = [ [[c[0], c[1]], [c[2], c[3]]]  for c in sorted(combinations) if not all([cc == -1 for cc in c]) and not all([cc == 0 for cc in c])and not all([cc == 1 for cc in c])]

# print(lis)

# Just rotation (but eivenvalues are kept real)
# ang = 60 # degree
# ang = ang/180*np.pi # degree
# rotation_matrix = np.array([
#     [np.cos(ang), -np.sin(ang)],
#     [np.sin(ang), np.cos(ang)]
# ])
# for i in range(len(lis)):
#     lis[i] = np.dot(rotation_matrix, lis[i])
    
# INPUT VECTORS ARE OF THE SAME LENGTH (NORM) OR NOT
useCircleInput=False

for lis in lislis:

    outheader = ["tl_mone", "tl_zero", "tl_one", "bl_mone", "bl_zero", "bl_one", "tr_mone", "tr_zero", "tr_one", "br_mone", "br_zero", "br_one",
                "det0", # col 12
                "orth1", # col 13
                "flipped", # col 14
                "rot0", # col 15
                 "matrix" # col 16
                ]
    outdata = []

    ind = 0
    for k in range(len(lis)):
        A = lis[k]

        print("------------------------------------------------------------------------------------------------------------------------------")
        print("")
        print("Transformation Matrix:")
        print("")
        print(f"[{np.round(A[0][0], 4)} {np.round(A[0][1], 4)}]")
        print(f"[{np.round(A[1][0], 4)} {np.round(A[1][1], 4)}]")
        print("")

        hor_rotate = np.degrees(np.arctan2(A[1][0], A[0][0]))
        ver_rotate = np.degrees(np.arctan2(A[1][1], A[0][1]))-90
        print("")
        print("=======================================================")
        print("    CHARACTERISTIC METRICS (RE-PARAMETERIZATION)")
        print("=======================================================")
        print("")
        squash_angle0 = ver_rotate-hor_rotate
        print(f"HORIZONTAL LINE ROTATION (atan of bl/tl):   {hor_rotate:.2f} degrees")
        print(f"VERTICAL LINE ROTATION (atan of br/tr):   {ver_rotate:.2f} degrees")
        print(f"  (new angle between 0 and 90 -> {squash_angle0+90:.2f} degrees)")

        stretch0 = norm_of([A[0][0], A[1][0]])
        stretch90 = norm_of([A[0][1], A[1][1]])
        stretch45 = norm_of([A[0][0]+A[0][1], A[1][0]+A[1][1]])/np.sqrt(2)
        stretch135 = norm_of([A[0][0]-A[0][1], A[1][0]-A[1][1]])/np.sqrt(2)
        print(f"STRETCH RATE OF 0 DEG LINE (|(tl bl)|):   {stretch0*100:.2f}%")
        print(f"STRETCH RATE OF 90 DEG LINE (|(tr br)|):   {stretch90*100:.2f}%")
        print(f"  (45 (|(tl+tr bl+br)|) -> {stretch45*100:.2f}%)")
        print(f"  (90 (|(tl-tr bl-br)|) -> {stretch135*100:.2f}%)")

        print("")
        print("=======================================================")
        print("")



        print("OTHER DERIVED METRICS:")
        print("")
        # output tab-delimited text
        outdata.append([0]*len(outheader))
        outdata[ind][16] = f"{A[0][0]:.3f} {A[0][1]:.3f} {A[1][0]:.3f} {A[1][1]:.3f}"
        for i in range(4):
            if A[int(i/3)%2][0] == -1:
                outdata[ind][i] = 1
            if A[int(i/3)%2][0] == 0:
                outdata[ind][i] = 1
            if A[int(i/3)%2][0] == 1:
                outdata[ind][i] = 1

        det = A[0][0]*A[1][1]-A[1][0]*A[0][1]
        print(f"Determinant (area of unit box) [tl*br - tr*bl]:    {det:.3f}")
        if det < 0.0001:
            outdata[ind][12] = 1
            A[0][0] = A[0][0]+0.00001

        orth = np.abs(det/(norm_of([A[0][0], A[1][0]])*norm_of([A[0][1], A[1][1]])))
        print(f"Orthognality [determinant / (|(tl bl)| * |(tr br)|)]:    {orth:.3f}")
        is_orthognal = False
        if orth > 0.9999:
            outdata[ind][13] = 1
            is_orthognal = True

        print("Is Flipped? [determinant < 0]:    ", (det < 0))
        if det < 0:
            outdata[ind][14] = 1

        ins_space=""
        if equals_more_or_less(det, 0):
            print("FYI for the case when the determinant approaches very closely to 0 ...")

            ins_space="   "

        eigenvalues, eigenvectors = np.linalg.eig(A)
        print(ins_space + f"Eigenvalues:    {eigenvalues[0]}, {eigenvalues[1]}")
        print(ins_space + "Eigen Vectors:   [", np.round(eigenvectors[0][0], 4), ",", np.round(eigenvectors[1][0], 4), "]", " [", np.round(eigenvectors[0][1], 4), ",", np.round(eigenvectors[1][1], 4), "]")
        en = [e for e in eigenvectors]    

        print("Is Rotated? (no fixed lines) [eigenvalues are complex numbers]:   ", (isinstance(eigenvalues[0], complex)))
        if isinstance(eigenvalues[0], complex):
            outdata[ind][15] = 1

        ind += 1

        print("")
        print("------------")
        print("")

        # Define the input vector
        v = []
        if useCircleInput:
            vvv = [[normalize_vector([s[0]+1,s[1]]), normalize_vector([s[0]+1,s[1]+1]), normalize_vector([s[0],s[1]+1]), normalize_vector([s[0],s[1]])] for s in [(0, 0), (1, 0), (0, 1), (1, 1), 
                                                                                      (0-2, 1),(1-2, 1), (0-2, 0), (1-2, 0), 
                                                                                      (0-2, 0-2), (1-2, 0-2), (0-2, 1-2), (1-2, 1-2),
                                                                                      (0, 1-2), (1, 1-2), (0, 0-2), (1, 0-2), 
                                                                                     ]]
        else:
            vvv = [[[s[0]+1,s[1]], [s[0]+1,s[1]+1], [s[0],s[1]+1], [s[0],s[1]]] for s in [(0, 0), (1, 0), (0, 1), (1, 1), 
                                                                                      (0-2, 1),(1-2, 1), (0-2, 0), (1-2, 0), 
                                                                                      (0-2, 0-2), (1-2, 0-2), (0-2, 1-2), (1-2, 1-2),
                                                                                      (0, 1-2), (1, 1-2), (0, 0-2), (1, 0-2), 
                                                                                     ]]
        for vvvv in vvv:
            v += vvvv

        print("")
        print("Scaled Input Vectors: ")
        print(convToListsWithRound(v, 4))

        print("")

        # Compute the output vector
        norms = []
        prevX = 0
        prevY = 0
        longest_output = max([norm_of(np.dot(A, v[i])) for i in range(len(v))])
        for i in range(len(v)):
            w = np.dot(A, v[i])
            if np.linalg.norm(w) > 0:
                norms.append(norm_of(w))

            # 
            # Plot the input and output vectors
            # 
            if not useCircleInput:
                plt.plot([prevX, w[0]], [prevY, w[1]], 'y', label='Output Vector', linewidth=1)
            prevX = w[0]
            prevY = w[1]
            cos_45_degrees = np.cos(np.radians(45))
            if v[i][1] != 0 and np.abs(v[i][0]/v[i][1])==1:
                if (v[i][0] == 1 and v[i][1] == 1) or (v[i][0] == 2 and v[i][1] == 2) or (useCircleInput and (equals_more_or_less(v[i][0], cos_45_degrees) and equals_more_or_less(v[i][1], cos_45_degrees)) or (equals_more_or_less(v[i][0], cos_45_degrees*2) and equals_more_or_less(v[i][1],cos_45_degrees*2))):
                    plt.plot([0, w[0]], [0, w[1]], 'orange', label='Output Vector', linewidth=5)
                elif (v[i][0] == -1 and v[i][1] == 1) or (v[i][0] == -2 and v[i][1] == 2) or (useCircleInput and (equals_more_or_less(v[i][0], -cos_45_degrees) and equals_more_or_less(v[i][1], cos_45_degrees)) or (equals_more_or_less(v[i][0], -cos_45_degrees*2) and equals_more_or_less(v[i][1],cos_45_degrees*2))):
                    plt.plot([0, w[0]], [0, w[1]], 'r', label='Output Vector', linewidth=5)
                else:
                    plt.plot([0, w[0]], [0, w[1]], 'pink', label='Output Vector', linewidth=2)
            else:
                plt.plot([0, w[0]], [0, w[1]], 'pink', label='Output Vector', linewidth=2)

            if not useCircleInput:
                posX = v[i][0]
                posY = v[i][1]
                plt.plot([0, 0], [v[i][1], 0], 'grey', label='Input Vector', linewidth=1)
                plt.plot([posX, 0], [posY, posY], 'grey', label='Input Vector', linewidth=1)
                plt.plot([posX, posX], [0, posY], 'grey', label='Input Vector', linewidth=1)
                plt.plot([0, posX], [0, 0], 'grey', label='Input Vector', linewidth=1)


        if det != 0 and not isinstance(eigenvalues[0], complex):
            plt.plot([-en[0][0]*eigenvalues[0], en[0][0]*eigenvalues[0]], [-en[1][0]*eigenvalues[0], en[1][0]*eigenvalues[0]], 'black', label='Eigen 1', linewidth=3)
            plt.plot([-en[0][0]*longest_output, en[0][0]*longest_output], [-en[1][0]*longest_output, en[1][0]*longest_output], 'green', label='Eigen 1', linewidth=0.5)
            plt.plot([-en[0][1]*eigenvalues[1], en[0][1]*eigenvalues[1]], [-en[1][1]*eigenvalues[1], en[1][1]*eigenvalues[1]], 'black', label='Eigen 2', linewidth=3)
            plt.plot([-en[0][1]*longest_output, en[0][1]*longest_output], [-en[1][1]*longest_output, en[1][1]*longest_output], 'green', label='Eigen 2', linewidth=0.5)


        plt.axis('equal')
        plt.show()

    print("")
    print("")
    print("")
    print("")
    outdata = [[str(element) for element in row] for row in outdata]
#     for r in ([outheader] + outdata):
#         print("\t".join(r))
```

# Two Point Conversion

Suppose you want to create a system of equations that conversts the given two points to specified two points.



One simple way may be to use conversion like below:



You can use more complex 4 parameter systems to enable.  You should need 4 parameters somehow to allow any 2 input points to any 2 output points.



The parameters p1, p2, p3, p4 can be obtained by simple algebra.
Algebraically, this can be resolved as there are 4 unknowns and 4 equations.  

$xo_1 = p1xi_1 + p2yi_1 $

$yo_1 = p3xi_1 + p4yi_1 $

$xo_2 = p1xi_2 + p2yi_2 $

$yo_2 = p3xi_2 + p4yi_2 $

In this case of A(4,6), B(5,1) - > C(-9,3), D(-2,8)

The equations become as follows:

$-9 = 4p1 + 6p2 $

$3 = 4p3 + 6p4 $

$-2 = 5p1 + 1p2 $

$8 = 5p3 + 1p4 $


The answers are: p1=-3/26, p2=-37/26, p3=45/26, p4-17/26 (verified in below Python program)

This two point conversion happens to make the net to net conversion.  
If the input points are given in the net format, the output data is also in the net format with the below characteristics
- lines remain lines
- parallel lines are kept parallel
- the distances between the lines remain in the constant ratio.

As sampled below:






This system of 4 parameter system of two point conversion is Linear Transformation, which is the heart of linear algabra.

```python
from sympy import symbols, Eq, linsolve

# Define the variables
p1, p2, p3, p4 = symbols('p1 p2 p3 p4')

# Define the system of equations
eq1 = Eq(4*p1 + 6*p2, -9)
eq2 = Eq(4*p3 + 6*p4, 3)
eq3 = Eq(5*p1 + 1*p2, -2)
eq4 = Eq(5*p3 + 1*p4, 8)

# Solve the system of equations
solution = linsolve([eq1, eq2, eq3, eq4], p1, p2, p3, p4)

# Display the solution
print("Solution:", solution)
```

```python
def linear_trans(T, A):
    [xi, yi] = A
#     -3/26, p2=-37/26, p3=45/26, p4-17/26
    xo = T[0][0]*xi + T[0][1]*yi
    yo = T[1][0]*xi + T[1][1]*yi
    return [xo, yo]


# T = [[-1. , 2.],[-2. ,  3.]]
T = [[-3/26 , -37/26],[45/26 ,  -17/26]]
print(linear_trans(T, [4, 6]))
print(linear_trans(T, [5, 1]))



```
